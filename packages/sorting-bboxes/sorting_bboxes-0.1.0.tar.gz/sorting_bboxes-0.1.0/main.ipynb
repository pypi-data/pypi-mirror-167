```python
import json
from PIL import Image, ImageDraw, ImageFont
import numpy as np
import cv2
```

```python
from sorting_bboxes import sort_by_line
```

```python
with open("./tests/data/0.json") as f:
    data = json.load(f)
```

```python
line_groups = sort_by_line(np.array(data["boxes"]), height=data["height"])
```

```python
img = Image.new("RGB", (data["width"], data["height"]), color="white")

draw = ImageDraw.Draw(img)
font = ImageFont.load_default()

for i, line in enumerate(line_groups):
    for j, box in enumerate(line):
        draw.polygon([tuple(point) for point in box], outline="blue")
        draw.text(tuple(box[0]), f"{i}\n{j}", align="left", fill="black", font=font)

img
```
